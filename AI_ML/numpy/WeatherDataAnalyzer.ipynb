```python
import numpy as np
```

```python
# step 1 : Generate random temp data
np.random.seed(0)
temps = np.random.randint(10,46, size=30)
print("Temp Data in (°C)")
print(temps)
```

```python
# average temperature
avg_temp = np.mean(temps)
avg_temp
```

```python
# hottest and coldest day
hottest_day = np.argmax(temps)
print(f"Hottest day: {hottest_day + 1}")
coldest_day = np.argmin(temps)
print(f"Coldest day: {coldest_day + 1}")
```

```python
# number of days above 35°C
hot_days = np.sum(temps > 35)
print(f"days above 35°C: {hot_days}")
```
