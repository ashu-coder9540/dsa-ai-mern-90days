```python
import numpy as np
```

```python
# 1. Create a marks dataset
# Create a 2D NumPy array of shape (10, 5)
# 10 students
# 5 subjects
# Marks between 0 and 100
np.random.seed(0)
marks = np.random.randint(0,101,(10,5))
marks
```

```python
# 2. Calculate total marks of each student
totalMarks = np.sum(marks, axis = 1)
totalMarks
```

```python
# 3. Calculate average marks of each subject
avgMarks = np.mean(marks, axis = 1)
avgMarks
```

```python
# 4. Identify the topper (highest total marks)
# first identify idx of highest total marks
topper_idx = np.argmax(totalMarks)
# now show total marks of topper
topperMarks = totalMarks[topper_idx]
topperMarks
```

```python
# 5. Find the subject with the highest average
# first calculate total avg of each sub
avgMarks
# identify idx of highest total avg
highestAvgIdx = np.argmax(avgMarks)
#now show highest avg value
highest_avg = avgMarks[highestAvgIdx]
highest_avg
```

```python
# 6. Assign grades
# A: 90 and above
# B: 75–89
# C: 60–74
# D: 40–59
# F: below 40
grades = np.where(avgMarks >= 90, 'A', 
              np.where(avgMarks >= 75, 'B',
              np.where(avgMarks >= 60, 'C',
              np.where(avgMarks >= 40, 'D',
              'F'))))
print("Average Marks:", avgMarks)
print("Grades:", grades)
```

```python
# 7. Normalize marks between 0 and 1
normalize = avgMarks/100
normalize
```
