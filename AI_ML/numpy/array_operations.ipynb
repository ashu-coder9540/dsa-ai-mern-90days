```python
import numpy as np
```

#### Array Operations

*Arithmetic Operations*

```python
a1 = np.arange(1,6)
a1
a2 = np.arange(6,11)
a2
```

```python
sum = a1 + a2
sum
```

```python
diff = a2 - a1
diff
```

```python
mul = a1 * a2
mul
```

```python
div = a1 / a2
div
```

```python
floor = a1 // a2
floor
```

```python
exp = a2 ** a1
exp
```

*Broadcasting*

```python
l = [10,20,30,40]
arr = np.array(l)
arr
```

```python
arr + 10
```

```python
arr2 = np.arange(1,26).reshape(5,5)
arr2
```

```python
arr2 + 10
```

```python
arr2 * 2
```

*Deep and Shallow copy*

```python
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice = slice * 10
slice
```

```python
a
```

```python
a[4:6] * 10
```

```python
a
b = a.copy()
```

```python
b[0] = 80
```

#### Advance Array Manipulation

*Stacking Array*

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

*Splitting Array*

```python
c = np.arange(16).reshape(4,4)
c
```

```python
np.hsplit(c, 4)
```

```python
a = np.vsplit(c, 4)
a
```

```python
for i in a: 
    print(i)
```
