```python
# 6. Slicing
# From arr = np.arange(1,21), get all even numbers using slicing.
arr = np.arange(1,21)
arr[1::2]
```

```python
# 7. Boolean Masking
# From arr = np.array([3,8,15,24,7,12]), filter values greater than 10.
arr = np.array([3,8,15,24,7,12])
arr[arr > 10]
```

```python
# 8. Fancy Indexing
# Given arr = np.array([10,20,30,40,50]), pick values at positions [0,3,4].
arr = np.array([10,20,30,40,50])
arr[[0,3,4]]
```

```python
# 9. Broadcasting
# Add a 1D array [1,2,3] to every row of a 3×3 matrix.
arr = np.arange(1,10)
arr_2d = arr.reshape((3,3))
arr_1d = np.array([1,2,3])
arr_2d + arr_1d
```

```python
# 10. Aggregations
# Find the mean, median and standard deviation of [1,2,3,4,5].
arr = np.array([1,2,3,4,5])
mean_arr = np.mean(arr)
median_arr = np.median(arr)
std_arr = np.std(arr)
std_arr
```

```python
# 11. Argmax / Argmin
# Given arr = np.array([4,99,2,50,7]), find the index of maximum value.
arr = np.array([4,99,2,50,7])
argMax = arr.argmax()
argMin = arr.argmin()
argMin
```

```python
# 12. Unique Values
# Find unique values and their counts from np.array([1,2,2,3,3,3,4]).
arr = np.array([1,2,2,3,3,3,4])
unique_Val, count = np.unique(arr, return_counts=True)
print(unique_Val)
print(count)
```

```python
# 13. Sorting
# Sort this array in descending order: np.array([5,1,7,3,9]).
arr = np.array([9,3,2,5,4,7,1,2])
res = np.sort(arr)[::-1]
res
```

```python
# 14. Concatenate
# Combine two arrays: a=[1,2,3] and b=[4,5,6] into one.
a=[1,2,3]
b=[4,5,6]
res = np.concatenate([a,b])
res
```

```python
# 15. Stack
# Stack two 1D arrays vertically to make a 2×3 matrix.
a=[1,2,3]
b=[4,5,6]
arr_23 = np.vstack((a,b))
```

```python
# 16. Where()
# Use np.where to replace negative values with 0 in
arr = np.array([4,-3,7,-1,9])
res = np.where(arr < 0, 0 , arr)
print(res)
```

```python
# 17. Linspace
# Create 10 numbers equally spaced between 0 and 5.
arr = np.linspace(0,5,10)
arr
```

```python
# 18. Matrix Multiplication
# Multiply a 2×3 matrix with a 3×2 matrix.
arr = np.arange(1,7)
arr_32 = arr.reshape((3,2))
print(arr_23)
print(arr_32)
```
