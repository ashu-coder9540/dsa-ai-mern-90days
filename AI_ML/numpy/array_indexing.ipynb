```python
import numpy as np
```

### NUMPY INDEXING AND SLICING OF VECTORS

```python
arr = np.arange(11,21)
arr
```

```python
arr_slice = arr[3::2]
arr_slice
```

```python
arr[4]
```

### NUMPY INDEXING AND SLICING OF MATRIX

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[5,4]
```

```python
arr[0:2,1:3]
```

```python
arr[3:6, 3:5]
```

```python
arr[:,2]
```

### BOOLEAN INDEXING

```python
arr = np.arange(11,21)
arr
```

```python
bool_idx = arr % 2 == 0
bool_idx
```

```python
arr = arr[bool_idx]
arr
```
